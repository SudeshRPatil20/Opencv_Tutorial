```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("D:\sudesh\All_about_DL\opencv_tutorial\color_thresholding\giraph.jpg", cv2.IMREAD_GRAYSCALE)

scocial_x=cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
scocial_y=cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

scocial_combine=cv2.magnitude(scocial_x, scocial_y)

#laplacian 

laplacian=cv2.Laplacian(image, cv2.CV_64F, ksize=3)
laplacian_abs=cv2.convertScaleAbs(laplacian)


#canny
canny_edges=cv2.Canny(image, 180, 200)

```

```python
titles=['original', 'sobal', 'laplacian', 'laplacian_abs', 'canny']
image_list=[image, scocial_combine, laplacian, laplacian_abs, canny_edges]

plt.figure(figsize=(20,16))
for i in range(len(titles)):
    plt.subplot(3, 2, i+1)
    plt.imshow(image_list[i], cmap='grey')
    plt.title(titles[i])
    plt.axis('off')
plt.show()
```
