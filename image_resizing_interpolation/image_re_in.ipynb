```python
#now we wolud do the image scaling resizing and interpolation
# scaling -> is changing the size of the image by changing its height and width
# resizing -> 
# interpolation -> 


import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_bgr=cv2.imread("./giraph.jpg")
```

```python
plt.imshow(image_bgr)
```

```python
img_rgb=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

# Image Scaling

```python
print(img_rgb.shape)
```

```python
# hear height is 1500 and width is 1000 so if i want to change the height and width
new_height, new_width=400 , 400
resizing_image=cv2.resize(img_rgb, (new_width, new_height))

plt.imshow(resizing_image)
```

# Image Scaling 

```python
image_scale=50
s_width=int(img_rgb.shape[1] * image_scale / 100)
s_height=int(img_rgb.shape[0] * image_scale / 100)

scaled_image=cv2.resize(img_rgb, (s_width, s_height), interpolation=cv2.INTER_AREA)

print(scaled_image.shape)
print("the shape of original is :", img_rgb.shape)

plt.imshow(scaled_image)
```

# interpolation

```python
resize_width=600
resize_height=600

method={
    'NEAREST':cv2.INTER_NEAREST,
    'LINEAR':cv2.INTER_LINEAR,
    'CUBIC':cv2.INTER_CUBIC,
    'AREA':cv2.INTER_AREA,
    'LANCZOS4':cv2.INTER_LANCZOS4
}

resize_images={}

for name, method in method.items():
    resize_images[name]=cv2.resize(img_rgb, (resize_width, resize_height), interpolation=method)
    cv2.imwrite(f"./{name}.jpg", resize_images[name])
    
plt.figure(figsize=(12, 8))


plt.subplot(2,3,1)
plt.imshow(img_rgb)
plt.title("Original_image")
plt.axis('off')

for i, (name, resize_images) in enumerate(resize_images.items(), start=2):
    plt.subplot(2,3,i)
    plt.imshow(resize_images)
    plt.title(name)
    plt.axis('off')
    
plt.tight_layout()
plt.show()
    
```
