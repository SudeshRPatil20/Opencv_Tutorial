```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
Image=cv2.imread("D:\sudesh\All_about_DL\opencv_tutorial\image_resizing_interpolation\giraph.jpg")
img_rgb=cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

```python
img_rgb.shape
```

```python
rows, cols, _= img_rgb.shape

input_points=np.float32([[50, 50], [cols - 50, 50], [50, rows - 50], [cols-50, rows-50]])

output_points=np.float32([[10, 100], [cols - 100, 50], [100, rows - 10], [cols - 50, rows - 100]])


''' 
Transforamtion metrix for prespective transformation

3x3

a, b, c
d, e, f
g, h, 1

a,b,c is responsible for horizontal scalling , tranformation, rotation 
d,e,f is responsible for vertical scalling , transformation, rotation
g, h, : Prespective transformation components responsible for depth and skew
'''

M = cv2.getPerspectiveTransform(input_points, output_points)

print(M)
```

```python
prespective_image=cv2.warpPerspective(img_rgb, M, (cols, rows))

plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.imshow(prespective_image)
plt.title("Prespective Image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(img_rgb)
plt.title("original_image")
plt.axis("off")

# below image black lines are blank imformation
```

```python
# Affine Transformation
input_points = np.float32([[50, 50], [200, 50], [50, 200]])
output_points = np.float32([[10, 100], [200, 50], [100, 250]])

M = cv2.getAffineTransform(input_points, output_points)

affine_images=cv2.warpAffine(img_rgb, M, (cols, rows))

plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("original_image")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(affine_images)
plt.title("AFFINE IMAGE")
plt.axis("off")
```
