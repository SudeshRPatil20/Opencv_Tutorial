```python
# Contrast limited adaptive Histogram Equilization (CLAHE)

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("D:\sudesh\All_about_DL\opencv_tutorial\color_thresholding\giraph.jpg")
```

```python

img_grey=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

equlized_image= cv2.equalizeHist(img_grey)

clahe=cv2.createCLAHE(clipLimit=2.0, tileGridSize=(4,4))

clahe_equilized_image=clahe.apply(img_grey)

plt.figure(figsize=(12,8))

plt.subplot(1,3,1)
plt.imshow(img_grey, cmap='grey')
plt.title("original_image")

plt.subplot(1,3,2)
plt.imshow(equlized_image, cmap='grey')
plt.title("Equilized_image")

plt.subplot(1,3,3)
plt.imshow(clahe_equilized_image, cmap='grey')
plt.title("Clahe_equilized_image")


plt.tight_layout()
plt.show()

```

```python
# now convering to yuv image

yuv_image = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)

yuv_image[:,:,0]= clahe.apply(yuv_image[:,:,0])

yuv_image_bgr= cv2.cvtColor(yuv_image, cv2.COLOR_YUV2BGR)

yuv_image_rgb=cv2.cvtColor(yuv_image_bgr, cv2.COLOR_BGR2RGB)

img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12,8))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("original_image")

plt.subplot(1,2,2)
plt.imshow(yuv_image_rgb)
plt.title("Clahe_Equilized_image")



plt.tight_layout()
plt.show()
```

#### Clahe is more advance version of then normal equilization here you are able to see that by clahe more clear and more sophesticated image is obtained which is cool and hlep in capturing imp features from image
