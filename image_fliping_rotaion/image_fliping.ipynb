```python
# image_fliping -> 
#image_rotaion ->
#image_croping -> 

import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_bgr=cv2.imread("D:\sudesh\All_about_DL\opencv_tutorial\color_thresholding\giraph.jpg")
```

# FLIP

```python
image_rgb=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
# plt.imshow(image_rgb)

flip_horizintal=cv2.flip(image_bgr, 1)
flip_vertical=cv2.flip(image_bgr, 0)
flip_both_side=cv2.flip(image_bgr, -1)

titles=['original_image', 'horizontal_image', 'vertical_image', 'both_side']
image_rotate=[image_rgb, 
              cv2.cvtColor(flip_horizintal, cv2.COLOR_BGR2RGB),
              cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB),
              cv2.cvtColor(flip_both_side, cv2.COLOR_BGR2RGB)]

plt.figure(figsize=(12,8))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(image_rotate[i])
    plt.title(titles[i])
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```

# Rotate

```python
plt.imshow(cv2.rotate(image_rgb, cv2.ROTATE_90_CLOCKWISE))
```

# Croping

```python
print(image_rgb.shape)
print(type(image_rgb))
```

```python
x_start, y_start=400 , 300
x_end, y_end=800, 900

croped_image=image_rotate[1][y_start:y_end, x_start:x_end]

plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image_rotate[1])
plt.title("original_image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(croped_image)
plt.title("croped_image")
plt.axis('off')


plt.tight_layout()
plt.show()
```
