```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("./giraph.jpg")
```

```python
plt.imshow(image)
```

```python
# image_rgb by defalt it take image as bgr so we convert image to rgb
img_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
img_hsv=cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(img_rgb)
```

remamber to check wether image is in rgb or hsv always check we do changes in hsv image

```python
#here i want to target white color so we do uper_bond and lower bond as white
lower_white=np.array([0,0,190])
uper_white=np.array([30,40,255])

mask=cv2.inRange(img_hsv, lower_white, uper_white)

plt.imshow(mask)
```

```python
black_hsv=np.uint8([[[120, 255, 0]]])
black_bgr=cv2.cvtColor(black_hsv, cv2.COLOR_HSV2BGR)[0][0]
```

```python
image_result=image.copy()
image_result[mask>0]=black_bgr
```

```python
img_result_rgb=cv2.cvtColor(image_result, cv2.COLOR_BGR2RGB)
```

```python
#plot
fig, axi=plt.subplots(1, 3, figsize=(12,9))


axi[0].imshow(img_rgb)
axi[0].set_title("Original Image")
axi[0].axis('off')

axi[1].imshow(mask)
axi[1].set_title("mask Image")
axi[1].axis('off')

axi[2].imshow(img_result_rgb)
axi[2].set_title("result Image")
axi[2].axis('off')


plt.tight_layout()
plt.show()
```
