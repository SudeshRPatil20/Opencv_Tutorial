```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("./filter.jpg")
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
#APPLYING FILLTER which will plays importtant role in cnn
x= -10
horizontal_filter=np.array([[-1, -1, -1],
                            [x, x, x],
                            [1, 1, 1]])

vertical_filter=np.array([[-1, x, 1],
                          [-1, x, 1],
                          [-1, x, 1]])

diagonal_filter=np.array([[x, -1, -1],
                            [1, x, -1],
                            [1, 1, x]])

horizontal_filter_image= cv2.filter2D(image, -1, horizontal_filter)
vertical_filter_image= cv2.filter2D(image, -1, vertical_filter)
diagonal_filter_image= cv2.filter2D(image, -1, diagonal_filter)


plt.figure(figsize=(10, 10))
plt.subplot(2,2,1)
plt.title("original_image")
plt.imshow(image)


plt.subplot(2,2,2)
plt.title("horizontal_scale_image")
plt.imshow(horizontal_filter_image)


plt.subplot(2,2,3)
plt.title("vertical_image")
plt.imshow(vertical_filter_image)


plt.subplot(2,2,4)
plt.title("diagonal_image")
plt.imshow(diagonal_filter_image)
```
