```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
height, width=500, 700
blue= (255, 127, 0)
cancas=np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv2.cvtColor(cancas, cv2.COLOR_BGR2RGB))

# first step :- created a canvas
```

# line Drawing


```python
cv2.line(cancas, (50,50), (640, 440), (255,255,255), 5)
plt.imshow(cv2.cvtColor(cancas, cv2.COLOR_BGR2RGB))
```

# Draw a Rectangle

```python
cv2.rectangle(cancas, (100, 100), (300, 300), (0,255,255), -1)

# if you use 3 it will thick if you use 2 it will more thick -1 will fill the rectangle by color you mention

plt.imshow(cv2.cvtColor(cancas, cv2.COLOR_BGR2RGB))
```

# Drawing CIrcle

```python
cv2.circle(cancas, (500, 150), 100, (0,0,0), -1)
plt.imshow(cv2.cvtColor(cancas, cv2.COLOR_BGR2RGB))
```

# Draw an polygon


```python
#poligon has point xn, yn, where n>2

points=np.array([[150, 400], [300, 400], [255, 300]], np.int32)
new_reshape_points=points.reshape((-1, 1, 2))

print(new_reshape_points)

cv2.polylines(cancas, [new_reshape_points], True, (255, 255, 255), 1)

plt.imshow(cv2.cvtColor(cancas, cv2.COLOR_BGR2RGB))
```
