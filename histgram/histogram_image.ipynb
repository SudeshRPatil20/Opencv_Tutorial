```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
Image=cv2.imread("D:\sudesh\All_about_DL\opencv_tutorial\color_thresholding\giraph.jpg")
img_rgb=cv2.cvtColor(Image, cv2.COLOR_BGR2RGB)
img_gray=cv2.cvtColor(Image, cv2.COLOR_BGR2GRAY)

```

```python
hist, bins =np.histogram(img_gray.flatten(), 256, [0, 256])

equilize_image=cv2.equalizeHist(img_gray)

equilize_hist, bins=np.histogram(equilize_image.flatten(), 256, [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2,2,1)
plt.imshow(img_gray, cmap='grey')
plt.title('Original_image')

plt.subplot(2,2,2)
plt.plot(hist, color='blue')
plt.title('hist_image')

plt.subplot(2,2,3)
plt.imshow(equilize_image, cmap='grey')
plt.title('equilize_image')

plt.subplot(2,2,4)
plt.plot(equilize_hist, color='green')
plt.title('equileze_hist_image')

plt.tight_layout()
plt.show()

# img_equi=cv2.cvtColor(equilize_image, cv2.COLOR_BGR2RGB)
```

```python
#YUV and LAB

img_yuv=cv2.cvtColor(Image, cv2.COLOR_BGR2YUV)

img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])

equilize_color_image = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

equilize_color_image_rgb=cv2.cvtColor(equilize_color_image, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title('Original_image')

plt.subplot(1,2,2)
plt.imshow(equilize_color_image_rgb)
plt.title('equilize_image')


plt.tight_layout()
plt.show()
```
